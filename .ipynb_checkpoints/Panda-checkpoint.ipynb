```python
#Create a file to practice the Panda 
df1 = pd.read_csv("data/store.csv") # TO input the file 
```

```python
# Methods of Data Frame 
df1.columns # To show columns 
df1.head()# To top 5 Show 
df1.tail()#  To last 5 Show
df1.describe()# TO gate mean , mininum ,max , mean all show
df1.info()# show all inforamtion show 
```

```python
df1
```

```python
# Clear Data 
#handling Missing value
df1.isnull() #Show the Null value of the data in TRUE or FALSE 
df1.isnull().sum() # Show the Total number of Null values with the col
df1.isnull().sum(axis = 1) # To show Axis Col value 
df1.dropna() # Drop that col is NUll value 
df1.dropna(axis = 1) # to Drop the value at col
df1.fillna("Non")# Fill the Value to in bracket 
```

```python
# Handle the Data Types And Date-time 
df1.dtypes # To show the Data Types in the Data 
df2 = df1.copy()# Create a copy of df1 
df2["age"] = df2["age"].fillna(0).astype("int64")
df2["age"] = df2["age"].astype("int64").copy() # To Change the Data Types 
df2.dtypes
```

```python
df2["city"] = df2["city"].fillna("Dhule") # Chenge the NaN value to "Dhule"
```

```python
df2 # show the change in Age to convert the data type
```

```python
# Hadling the  String 
df2["city"] # show the col
df2["city"].str.upper() # convert in the Uper String 
df2["city"].str.lower() # convert in the lower String 
df2["city"].str.capitalize() # convert in the capitalize String 
df2["city"].str.split() # covert in Up doen step 
df2["city"].str.strip() # Extra space remove 
df2["city"].str.contains("Mumbai") # This value is present or not show in TRUE OR FALSE 
```

```python
df2
```

```python
#Feature Engineering 
df2["total_marks"] = df2["marks"].apply(lambda  x : "Good" if x >= 80 else "Better Try")
df2 # In this line  The Lambda funtion we use in the apply() method use 
```

```python
print(df2.columns)
```

```python
# map() fuction  
con = { "Good" : "👍" , "Better Try" : "👌" }
df2["total_marks"] = df2["total_marks"].map(con)
# To change the Value of the  Data 
```

```python
# assign Fuction 
df2.assign(new_marks = df2["marks"] * 1.1 )
```

```python
#replace Value 
df2["age"] = df2["age"].replace(0 , 18)
```

```python
#Transforming Data
df2 
#Change the Columns names 
df2.columns = ["Name", "Age","Marks","City","Grade_of_marks",] # change All col Names 
df2.rename(columns={"Name" : "Naav"}) # Inthis line we change the col Name 
```

```python
#Shor values
df2
df2.sort_values("Marks")# The Sort the value inthe Marks Column 
df2.sort_values("Marks" , ascending=False)# sort but not ascending Order
sort_val = df2.sort_values(["Marks" ,"Age"])# sort in two columns 
sort_val.sort_index()# sort the Index 
```

```python
df2 = df2.sort_values("Marks") # Short the Date on Marks
df2 = df2.reset_index(drop=True) # Set the Data in indexc form 
```

```python
 # To Create the CSV File to short the DATA  
df2.to_csv("short_data.csv")
```

```python
df2["Marks"] = df2["Marks"].fillna(36)
df2
```

```python
# Create a Col TO Classfiy the Subject Of the Students 
df2["Subject"] = df2["Marks"].apply(lambda x : "English" if x in range(0, 40 )
                                    else ("Math" if x in range(40 ,70 ) 
                                    else "Science"))
df2
```

```python
# Grouping And Aggrigation of Data 
df2.groupby("City")["Marks"].max() # To Max Number to the City 
df2.groupby("City")["Marks"].sum()# The Sum of Numbers of the marks 
df2.groupby("Subject")["Marks"].mean() # Show the Mean of the Subject Marks 
df2.groupby("City")["Marks"].agg(["max","mean","min"])# All type of Marks Calculate 
df2.groupby("City")["Marks"].agg(Max_MArks = "max",Mean_Marks = "mean",Min_Marks = "min") # Assigne the name of the col 
```

```python
df2.groupby("City").agg(
    Min_Marks = ("Marks","min"),
    Min_Age = ("Age","min")
)
```

```python
# Create a DataFream 
df4 = pd.DataFrame({
    "Country" : ["India","USA","India","USA"],
    "Year" : [2020,2022,2020,2022],
    "Sales" :[100,120,100,120],
    "Profit" : [20,30,10,20],
})   
```

```python
df4
```

```python
#Melt Fuction = Wide Format to long Formate  Convert 
melt_df = df4.melt(
    id_vars=["Country","Year"],     # In the col Country And Year Name is present 
    value_vars=["Sales","Profit"],  #In the Col Sales And Profit present in one Col or Marg
    var_name="Matrix",              # Assing the Name of  col that sales and profit marge 
    value_name="Value"              # To Create a col to All Values Store in the Col
) 
```
